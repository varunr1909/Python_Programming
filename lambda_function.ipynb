```python
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x**2, numbers))
print(squared_numbers)
```

```python
nums = [10, 15, 21, 33, 42, 50]
even_num=list(filter(lambda x:x%2==0,nums))
print(even_num)
```

```python
students = [('John', 82), ('Alice', 91), ('Bob', 75)]
sorted = list(sorted(students, key=lambda x: x[1], reverse=True))
print(sorted)
```
