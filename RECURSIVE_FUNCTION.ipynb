```python
#factorial with recursive function
def factorial(n):
  if(n==0 or n==1):
    return 1
  else:
    return n * factorial(n-1)
print(factorial(4))
```

```python
#sum of natural number with recursive function
def natural(n):
  if (n==0 or n==1):
    return 1
  else:
    return n+ natural(n-1)
print(natural(5))
```

```python
#fibbonacci sequence with recursive function
def fibb(n):
  if (n==0 or n== 1):
    return 1
  else:
    return fibb(n-1)+fibb(n-2)
print(fibb(4))
```

```python
def power(base, exp):
    if exp == 0:
        return 1
    else:
        return base * power(base, exp - 1)

# Example usage
print(power(2, 3))  # Output: 8
```

```python
#palindrome code with recursive function
def countdigit(n):
  if(n==0):
    return 0
  else:
    return 1+countdigit(n//10)
print(countdigit(12345))
```

```python
def palindrome(str):
  if (str == str[::-1]):
    return True
  else:
    return False
print(palindrome("madam"))
```
